```python
# -*- coding:utf8 -*-
import os
import csv
import pandas as pd
import time
path_train = "../data/dm/train.csv"  # 训练文件
path_test = "../data/dm/test.csv"  # 测试文件
path_test_out = "model/"

data = pd.read_csv(path_train)
a = data.groupby(['TERMINALNO',"TRIP_ID"])
c =list(a.groups)

print(data.shape)
# for tmp in c:
#     print(tmp)

#a.get_group((1,2))
b = a.get_group((1,3))
print(b)
print('len: ----------------------------',len(b))
print(b.values[1][4])

print('len: ----------------------------',len(b.values[:]))
#print(a.size())

```

```python
start_time = time.time()

with (open(os.path.join(path_test_out, "tmp.csv"), mode="w")) as outer:
        writer = csv.writer(outer)
        writer.writerow(["TERMINALNO", "TIME", "TRIP_ID", "LONGITUDE", 
                         "LATITUDE", "DIRECTION", "HEIGHT", "SPEED",
                        "CALLSTATE", "Y"])
        for tmp in c:
            b = a.get_group(tmp)
            if len(b)==1:
                continue
            for line in range(len(b)-1):
                writer.writerow([b.values[line][0],b.values[line][1],
                                 b.values[line][2],
                                 b.values[line+1][3] - b.values[line][3],
                                 b.values[line+1][4] - b.values[line][4],
                                 b.values[line+1][5] - b.values[line][5],
                                 b.values[line+1][6] - b.values[line][6],
                                 b.values[line+1][7] - b.values[line][7],
                                 b.values[line][8], b.values[line][9]])
end_time = time.time()
time_elapsed =end_time - start_time
print('save in {:.0f}m {:.0f}s'.format(time_elapsed // 60, time_elapsed % 60))
```

```python
train_data = pd.read_csv(path_train)
train_input = []
a = train_data.groupby(['TERMINALNO',"TRIP_ID"])
c =list(a.groups)
for tmp in c:
    b = a.get_group(tmp)
    if len(b)==1:
        continue
    for line in range(len(b)-1):
        index = [5,6,7,8,9]
        for k in index:
            if k < 8:
                train_input.append(b.values[line+1][index] - b.values[line][index])
            else:
                train_input.append(b.values[line][index])
            
print('len of train :',len(train_input))                
```
